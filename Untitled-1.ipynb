```python
import os
import random
import numpy as np
import pandas as pd

RANDOM_SEED = 42
os.environ['PYTHONHASHSEED'] = str(RANDOM_SEED)
random.seed(RANDOM_SEED)
np.random.seed(RANDOM_SEED)

from sklearn.model_selection import train_test_split, GridSearchCV, StratifiedKFold, cross_val_predict
from sklearn.preprocessing import StandardScaler, MinMaxScaler, LabelEncoder
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
from sklearn.metrics import (accuracy_score, precision_score, recall_score, f1_score,
                             roc_auc_score, roc_curve, confusion_matrix, classification_report)
import matplotlib.pyplot as plt
import seaborn as sns

# Models
from sklearn.naive_bayes import GaussianNB, MultinomialNB, ComplementNB, BernoulliNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier

# TensorFlow for NN
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.callbacks import EarlyStopping, TensorBoard

DATA_PATH = "spotify_churn_dataset.csv"
TARGET_COL = "is_churned" 
```

Загрузка датасета spotify_churn_dataset.csv с помощью pd.read_csv(). Сразу выводятся размеры таблицы (8000 строк 12 столбцов) и первые пять строк. Также вызываются методы .info() и .describe(include='all').T, которые показывают типы колонок, наличие пропусков и основные статистики - средние, медианы, уникальные значения и т.д.

Из .info() видно, что пропусков в данных нет (все 8000 значений в каждом столбце - non-null). А из .describe() видно, что целевой признак is_churned - бинарный (0 или 1), а признаки как числовые (age, listening_time и др.), так и категориальные (gender, country, subscription_type, device_type).

```python
# Загрузка
df = pd.read_csv(DATA_PATH)
print("Rows, cols:", df.shape)
display(df.head())
display(df.info())
display(df.describe(include='all').T)
```

 Если тип целевой переменной строковый, я преобразую её в числа 0 и 1. После этого я определяю категориальные признаки, выбирая столбцы строкового или категориального типа. Если целевой столбец оказался в этом списке, я удаляю его, чтобы не кодировать метку класса. В результате вывод показывает четыре категориальных признака, которые действительно содержат текстовые значения и потребуют последующего кодирования.


```python
# Если категориальные признаки — кодируем:
cat_cols = df.select_dtypes(include=['object', 'category']).columns.tolist()
# исключаем целевой, если он строчный
if TARGET_COL in cat_cols:
    cat_cols.remove(TARGET_COL)

print("categorical columns:", cat_cols)
```

В этой ячейке я удаляю колонку user_id, потому что она не несёт полезной информации. Далее я разделяю данные на матрицу признаков и целевую переменную, приводя целевую колонку к целочисленному типу. Затем я определяю список числовых признаков, выбирая все столбцы с числовым типом данных, а категориальные признаки уже были получены ранее. Итоговый вывод показывает, что в данных шесть числовых и четыре категориальных признака, что подтверждает правильность разбиения и соответствует структуре датасета.

```python
# Разделяем признаки и цель
# Удаляем ненужные колонки
df = df.drop(columns=['user_id'])
X = df.drop(columns=[TARGET_COL])
y = df[TARGET_COL].astype(int)

numeric_cols = X.select_dtypes(include=[np.number]).columns.tolist()
print("numeric:", len(numeric_cols), "categorical:", len(cat_cols))
```

В этом блоке я создаю препроцессор, который будет автоматически обрабатывать числовые и категориальные признаки перед подачей в модель. Для числовых колонок я настраиваю пайплайн, где сначала пропуски заменяются средним значением, а затем данные масштабируются при помощи MinMaxScaler, чтобы все значения были приведены к единому диапазону и не содержали отрицательных величин. Для категориальных признаков я задаю замену пропусков наиболее частым значением, после чего выполняю one-hot кодирование с игнорированием неизвестных категорий, что защищает модель от ошибок при встрече новых значений. Все эти преобразования объединяются в ColumnTransformer, который чётко разделяет числовые и категориальные переменные и применяет к ним соответствующие пайплайны.


```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import MinMaxScaler

numeric_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='mean')),
    ('scaler', StandardScaler())
])

categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('ohe', OneHotEncoder(handle_unknown='ignore'))
])

preprocessor = ColumnTransformer(transformers=[
    ('num', Pipeline([
        ('imputer', SimpleImputer(strategy='mean')),
        ('scaler', MinMaxScaler())  # не даёт отрицательных значений
    ]), numeric_cols),
    ('cat', Pipeline([
        ('imputer', SimpleImputer(strategy='most_frequent')),
        ('ohe', OneHotEncoder(handle_unknown='ignore'))
    ]), cat_cols)
])
```

В этой ячейке я делю данные на обучающую и тестовую выборки, сохраняя одинаковое распределение классов благодаря параметру stratify. После разбиения я применяю препроцессор: сначала обучаю его на тренировочных данных, чтобы он вычислил параметры масштабирования и категории для one-hot кодирования, а затем использую те же параметры для преобразования тестовой выборки.  Далее я вычисляю веса классов с помощью balanced, чтобы компенсировать дисбаланс. Итоговый словарь весов показывает, насколько сильнее модель должна штрафовать ошибки на меньшинстве по сравнению с большинством.

```python
# Разделение данных
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=RANDOM_SEED
)

# Применяем препроцессор к train
X_train_proc = preprocessor.fit_transform(X_train)
X_test_proc = preprocessor.transform(X_test)

# Подсчёт весов классов
from sklearn.utils.class_weight import compute_class_weight
class_weights = compute_class_weight('balanced', classes=np.unique(y_train), y=y_train)
class_weights = dict(zip(np.unique(y_train), class_weights))
```

Функция для оценки моделей. Сначала модель обучается на уже преобразованных тренировочных данных, после чего я получаю предсказания на тестовой выборке. Для расчёта ROC AUC мне нужны вероятности положительного класса, поэтому я проверяю, умеет ли модель выдавать predict_proba. Если такой метод есть, я беру второй столбец вероятностей, соответствующий классу 1. Если его нет, я пытаюсь получить решение через decision_function, и только если это тоже невозможно, использую сами предсказания как замену.

После получения предсказаний и вероятностей я рассчитываю набор стандартных метрик качества, включающий accuracy, precision, recall, f1 и auc. Для избежания ошибок при делении на ноль я задаю zero_division=0. Все значения собираются в словарь, который возвращается вместе с обученной моделью, предсказаниями и оценками.

```python
# обучение модели, предсказания, вычисление метрик и ROC
def evaluate_model(model, X_train_proc, X_test_proc, y_train, y_test):
    model.fit(X_train_proc, y_train)
    y_pred = model.predict(X_test_proc)
    
    if hasattr(model, "predict_proba"):
        y_score = model.predict_proba(X_test_proc)[:, 1]
    else:
        try:
            y_score = model.decision_function(X_test_proc)
        except:
            y_score = y_pred

    metrics = {
        'accuracy': accuracy_score(y_test, y_pred),
        'precision': precision_score(y_test, y_pred, zero_division=0),
        'recall': recall_score(y_test, y_pred, zero_division=0),
        'f1': f1_score(y_test, y_pred, zero_division=0),
        'auc': roc_auc_score(y_test, y_score)
    }
    return model, y_pred, y_score, metrics

```

В этом блоке я формирую словарь с набором моделей, которые затем буду последовательно обучать и сравнивать. Каждой модели я присваиваю строковый ключ, чтобы удобно обращаться к ней по имени при переборе. В набор входят разные алгоритмы: несколько разновидностей наивного Байеса, дерево решений, линейный дискриминантный анализ, SVM с включённым расчётом вероятностей и учетом дисбаланса классов, а также метод ближайших соседей. Это позволяет оценить, как алгоритмы разных типов ведут себя на этих данных, и получить представление о том, какие подходы работают лучше всего.


```python
# Список моделей
models = {
    'GaussianNB': GaussianNB(),
    'MultinomialNB': MultinomialNB(),
    'ComplementNB': ComplementNB(),
    'BernoulliNB': BernoulliNB(),
    'DecisionTree': DecisionTreeClassifier(class_weight='balanced', random_state=RANDOM_SEED),
    'LDA': LinearDiscriminantAnalysis(),
    'SVM': SVC(class_weight='balanced', probability=True, random_state=RANDOM_SEED),
    'KNN': KNeighborsClassifier()
}
```

Для каждой модели я создаю отдельный пайплайн, в котором сначала применяется заранее подготовленный препроцессор, а затем запускается сам классификатор. Пайплайн обучается на тренировочной выборке в исходном формате, потому что все необходимые преобразования выполняются внутри него. После обучения я получаю предсказания на тестовых данных и извлекаю вероятности положительного класса, если модель умеет их выдавать, либо использую решение функции, а при необходимости - сами метки.

Для каждой модели я вычисляю ключевые метрики: accuracy, precision, recall, f1 и auc. Затем сохраняю результаты в словарь и дополнительно сохраняю сам пайплайн, чтобы позже не переобучать его повторно. Вывод показывает, что большинство наивных байесов и линейный дискриминантный анализ фактически игнорируют положительный класс, что видно по нулевым значениям precision, recall и f1, хотя accuracy у них остаётся высокой из-за доминирования отрицательного класса. Некоторые алгоритмы, такие как ComplementNB, SVM и KNN, дают ненулевые показатели по положительному классу, но в целом их качество остаётся ограниченным. Значения AUC также демонстрируют, что модели пока слабо различают классы, что является ожидаемым результатом на базовой стадии без подбора гиперпараметров.

```python
results = {}
pipes = {}
for name, model in models.items():
    try:
        # Оборачиваем модель в пайплайн с препроцессором для последующего использования
        pipe = Pipeline(steps=[('preproc', preprocessor), ('clf', model)])
        pipe.fit(X_train, y_train)  # обучаем на исходных данных (preprocessor внутри)

        # Предсказания на тесте
        y_pred = pipe.predict(X_test)
        if hasattr(pipe, "predict_proba"):
            y_score = pipe.predict_proba(X_test)[:, 1]
        else:
            try:
                y_score = pipe.decision_function(X_test)
            except:
                y_score = y_pred

        metrics = {
            'accuracy': accuracy_score(y_test, y_pred),
            'precision': precision_score(y_test, y_pred, zero_division=0),
            'recall': recall_score(y_test, y_pred, zero_division=0),
            'f1': f1_score(y_test, y_pred, zero_division=0),
            'auc': roc_auc_score(y_test, y_score)
        }
        results[name] = metrics
        pipes[name] = pipe
        print(f"{name}: {metrics}")
    except Exception as e:
        print(f"{name} failed: {e}")
```

В этой ячейке я преобразую словарь с метриками в таблицу DataFrame, чтобы удобнее сравнивать модели. После создания таблицы я сортирую её по f1-мере в убывающем порядке, поскольку f1 позволяет оценить баланс между полнотой и точностью для несбалансированных данных. Вывод показывает, что среди всех алгоритмов лучшие результаты по f1 демонстрируют SVM и ComplementNB, хотя их показатели остаются довольно низкими. Большинство моделей имеют нулевые значения precision и recall, что подтверждает их склонность полностью предсказывать большинство классов одним образом.

```python
# Таблица результатов
res_df = pd.DataFrame(results).T
display(res_df.sort_values(by='f1', ascending=False))
```

В этой ячейке я визуализирую ROC-кривые для всех обученных моделей, чтобы оценить их способность различать классы при разных порогах. Сначала для каждого пайплайна я получаю вероятности положительного класса либо через predict_proba, либо через decision_function, если первый метод недоступен. Затем вычисляю пары значений FPR и TPR и добавляю соответствующую линию на график вместе с подписью, где указываю имя модели и значение AUC. На графике также есть диагональная линия, отображающая поведение случайного классификатора, что служит точкой сравнения. 

Визуальный результат показывает, что все кривые располагаются очень близко к диагонали, что подтверждает низкую способность моделей отделять классы друг от друга. Кривые слегка различаются, и у некоторых алгоритмов AUC оказывается немного выше, как у KNN, но в целом общее качество остаётся примерно одинаковым и сравнительно слабым.

```python
# ROC-кривые всех моделей
plt.figure(figsize=(8,6))
for name, pipe in pipes.items():
    try:
        y_score = pipe.predict_proba(X_test)[:,1]
    except:
        try:
            y_score = pipe.decision_function(X_test)
        except:
            continue
    fpr, tpr, _ = roc_curve(y_test, y_score)
    plt.plot(fpr, tpr, label=f"{name} (AUC={roc_auc_score(y_test,y_score):.3f})")
plt.plot([0,1],[0,1],'k--', alpha=0.4)
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curves")
plt.legend(loc='lower right')
plt.grid(True)
plt.show()
```

В этой ячейке я подготавливаю всё необходимое для подбора гиперпараметров моделей. Сначала я создаю объект StratifiedKFold, чтобы кросс-валидация проходила с сохранением исходного распределения классов в каждом сплите. После этого я формирую словарь с сетками параметров для каждой модели. Для деревьев решений, SVM и KNN я задаю несколько значений ключевых гиперпараметров, которые чаще всего влияют на качество классификации. Для наивных байесов я указываю возможные значения параметра сглаживания alpha там, где он используется, а для GaussianNB оставляю пустой словарь, поскольку у него нет параметров, влияющих на процесс обучения. Для LDA я фиксирую единственный доступный в данном контексте solver.

```python
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=RANDOM_SEED)

param_grids = {
    'DecisionTree': {
        'clf__max_depth': [3, 5, 10, None],
        'clf__min_samples_split': [2, 5, 10],
        'clf__criterion': ['gini', 'entropy']
    },
    'SVM': {
        'clf__C': [0.1, 1, 10],
        'clf__kernel': ['rbf', 'linear'],
        'clf__gamma': ['scale', 'auto']
    },
    'KNN': {
        'clf__n_neighbors': [3,5,7,9],
        'clf__weights': ['uniform','distance']
    },
    'GaussianNB': {
        # у GaussianNB нет гиперпараметров
    },
    'MultinomialNB': {
        'clf__alpha': [0.01, 0.1, 1.0]
    },
    'BernoulliNB': {
        'clf__alpha': [0.01, 0.1, 1.0]
    },
    'ComplementNB': {
        'clf__alpha': [0.01, 0.1, 1.0]
    },
    'LDA': {
        'clf__solver': ['svd']
    }
}
```

В этой ячейке я выполняю перебор гиперпараметров для выбранных моделей, используя сетки параметров, подготовленные ранее. Для каждой модели я создаю пайплайн, который объединяет препроцессор и сам классификатор, чтобы GridSearch мог корректно применять все этапы обработки на каждом сплите. Если для модели существует сетка гиперпараметров, я запускаю GridSearchCV с оптимизацией по f1-мере и стратифицированной кросс-валидацией. Обучение проходит на тренировочной выборке, и по завершении я вывожу лучшие найденные параметры и средний показатель f1. Если сетка оказалась пустой, как, например, у LDA, я просто обучаю пайплайн без перебора. 

Результаты показывают, что дерево решений после настройки улучшает показатель f1 и выбирает относительно небольшую глубину с критерием энтропии. SVM также находит неплохую конфигурацию с rbf-ядром и автоматическим gamma, хотя итоговое качество остаётся умеренным. KNN выбирает три ближайших соседа с весами по расстоянию, что соответствует его лучшей стратегии на этих данных. Из наивных байесов лучший результат демонстрирует ComplementNB, тогда как остальные продолжают давать нулевую f1.

```python
# Выполним GridSearch для нескольких моделей
best_models = {}
for name in ['DecisionTree','SVM','KNN','MultinomialNB','BernoulliNB','ComplementNB','LDA']:
    model = models[name]
    grid = param_grids.get(name, {})
    pipe = Pipeline(steps=[('preproc', preprocessor), ('clf', model)])
    if grid:
        gs = GridSearchCV(pipe, param_grid=grid, scoring='f1', cv=cv, n_jobs=-1, verbose=1)
        gs.fit(X_train, y_train)
        print(f"{name} best params: {gs.best_params_}, best f1: {gs.best_score_:.4f}")
        best_models[name] = gs.best_estimator_
    else:
        pipe.fit(X_train, y_train)
        best_models[name] = pipe
```

В этом блоке происходит окончательная оценка всех моделей, для которых ранее был выполнен подбор гиперпараметров с помощью GridSearchCV. Каждая модель (в виде пайплайна, включающего препроцессинг и сам алгоритм) применяется к тестовой выборке: сначала делаются предсказания меток, затем — оценки вероятностей (или, при их отсутствии, оценки через decision function), чтобы можно было посчитать AUC-ROC. После этого для каждой модели вычисляются ключевые метрики качества: accuracy, precision, recall, F1-score и AUC.

Результаты собираются в словарь и преобразуются в DataFrame, отсортированный по F1-score — это особенно важно, так как целевой класс несбалансирован (лишь около 26% пользователей ушли), и F1 учитывает оба класса сильнее, чем обычная точность. Видно, что ComplementNB показывает наилучший F1 (примерно 0.35), несмотря на скромную точность, потому что он лучше всего находит положительные случаи (recall ≈ 0.48). При этом такие модели, как MultinomialNB, BernoulliNB и LDA, формально имеют высокую accuracy (~0.74), но полностью игнорируют класс оттока — их precision и recall равны нулю, что означает, что они всегда предсказывают «не ушёл». Это типичный пример обманчивости accuracy при дисбалансе классов. В итоге становится ясно, что даже лучшие из классических моделей справляются с задачей слабо, и F1 в районе 0.35 — это практический предел на данном наборе признаков.

```python
# Оценка лучших моделей
final_results = {}
for name, pipe in best_models.items():
    y_pred = pipe.predict(X_test)
    try:
        y_score = pipe.predict_proba(X_test)[:,1]
    except:
        try:
            y_score = pipe.decision_function(X_test)
        except:
            y_score = y_pred
    metrics = {
        'accuracy': accuracy_score(y_test, y_pred),
        'precision': precision_score(y_test, y_pred, zero_division=0),
        'recall': recall_score(y_test, y_pred, zero_division=0),
        'f1': f1_score(y_test, y_pred, zero_division=0),
        'auc': roc_auc_score(y_test, y_score)
    }
    final_results[name] = metrics
pd.DataFrame(final_results).T.sort_values(by='f1', ascending=False)
```

```python
import tempfile
import datetime
from tensorflow.keras.callbacks import TensorBoard, EarlyStopping
import os

# Преобразуем данные
X_train_proc = preprocessor.fit_transform(X_train)
X_test_proc = preprocessor.transform(X_test)

# Определяем входную размерность
input_shape = X_train_proc.shape[1]
tf.random.set_seed(RANDOM_SEED)

# Функция построения модели
def build_model(hidden_units=[64, 32], lr=1e-3, dropout=0.2):
    model = keras.Sequential()
    model.add(layers.Input(shape=(input_shape,)))
    for units in hidden_units:
        model.add(layers.Dense(units, activation='relu'))
        if dropout > 0:
            model.add(layers.Dropout(dropout))
    model.add(layers.Dense(1, activation='sigmoid'))
    model.compile(
        optimizer=keras.optimizers.Adam(learning_rate=lr),
        loss='binary_crossentropy',
        metrics=['accuracy', keras.metrics.Precision(), keras.metrics.Recall(), keras.metrics.AUC(name='auc')]
    )
    return model


# Создание модели
model = build_model(hidden_units=[128, 64], lr=1e-3, dropout=0.3)
model.summary()


# Уникальная папка для TensorBoard
log_dir = os.path.join(tempfile.gettempdir(), "tf_logs", datetime.datetime.now().strftime("%Y%m%d-%H%M%S"))

# Создаём родительскую папку
os.makedirs(log_dir, exist_ok=True)

tensorboard_callback = TensorBoard(log_dir=log_dir, histogram_freq=1)
early_stop = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True, verbose=1)

history = model.fit(
    X_train_proc, y_train,
    validation_split=0.2,
    epochs=100,
    batch_size=32,
    class_weight=class_weights,
    callbacks=[tensorboard_callback, early_stop],
    verbose=1
)


# === Оценка на тестовой выборке ===
y_pred_proba = model.predict(X_test_proc)
y_pred = (y_pred_proba > 0.5).astype(int)


print("Логи сохраняются в:", log_dir)

print("\nClassification Report:")
print(classification_report(y_test, y_pred))
print("AUC:", roc_auc_score(y_test, y_pred_proba))
```

В этом блоке реализуется поиск лучших гиперпараметров нейронной сети с помощью кросс-валидации и последующее обучение финальной модели на всех данных. В отличие от предыдущего подхода с отложенной тестовой выборкой, здесь используется StratifiedKFold (5 фолдов) для более надёжной оценки качества каждой конфигурации модели. Это позволяет избежать переобучения на одном разбиении и получить более устойчивую оценку.

Сначала данные полностью препроцессируются с помощью ColumnTransformer (численные признаки масштабируются через StandardScaler, категориальные - кодируются OneHotEncoder). Затем по сетке гиперпараметров (ParameterGrid) перебираются комбинации архитектуры сети (hidden_units), скорости обучения (lr), коэффициента dropout и других параметров. Для каждой комбинации модель обучается 5 раз на разных фолдах, и в качестве метрики качества сохраняется средняя accuracy по кросс-валидации.

Наибольшую среднюю точность (0.5772) показала конфигурация:
hidden_units=[128, 64, 32], dropout=0.3, lr=0.0001.

Эти параметры используются для обучения финальной модели на всех данных (X_proc, y). В процессе обучения (50 эпох) модель постепенно улучшает свои показатели: к концу обучения accuracy достигает 0.53, AUC - 0.55, а recall стабильно держится около 0.54, что говорит о способности находить ушедших пользователей, хотя и с низкой точностью предсказаний (precision = 0.29).

После завершения обучения модель применяется к всем данным (без разделения на train/test), и её accuracy на всех 8000 записях оказывается 0.6019. 

```python
from sklearn.model_selection import StratifiedKFold, cross_val_score
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.utils.class_weight import compute_class_weight
import tensorflow as tf
from tensorflow.keras import layers, models, optimizers, callbacks
import numpy as np
import pandas as pd
import os
import tempfile
import datetime

# Фиксируем случайность
RANDOM_SEED = 42
np.random.seed(RANDOM_SEED)
tf.random.set_seed(RANDOM_SEED)

# Загрузка данных
DATA_PATH = "spotify_churn_dataset.csv"
TARGET_COL = "is_churned"
df = pd.read_csv(DATA_PATH)

# Разделение X/y
df = df.drop(columns=['user_id'])
X = df.drop(columns=[TARGET_COL])
y = df[TARGET_COL].astype(int)

# Категориальные и числовые признаки
cat_cols = X.select_dtypes(include=['object']).columns.tolist()
num_cols = X.select_dtypes(include=[np.number]).columns.tolist()

# Препроцессор
numeric_transformer = Pipeline([
    ('imputer', SimpleImputer(strategy='mean')),
    ('scaler', StandardScaler())
])

categorical_transformer = Pipeline([
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('ohe', OneHotEncoder(handle_unknown='ignore'))
])

preprocessor = ColumnTransformer([
    ('num', numeric_transformer, num_cols),
    ('cat', categorical_transformer, cat_cols)
])

# Классовые веса
class_weights = compute_class_weight('balanced', classes=np.unique(y), y=y)
class_weights = dict(zip(np.unique(y), class_weights))

def build_model(hidden_units=[64,32], lr=1e-3, dropout=0.2):
    model = models.Sequential()
    model.add(layers.Input(shape=(X_train_proc.shape[1],)))
    for units in hidden_units:
        model.add(layers.Dense(units, activation='relu'))
        if dropout > 0:
            model.add(layers.Dropout(dropout))
    model.add(layers.Dense(1, activation='sigmoid'))
    model.compile(
        optimizer=optimizers.Adam(learning_rate=lr),
        loss='binary_crossentropy',
        metrics=['accuracy', tf.keras.metrics.Precision(), tf.keras.metrics.Recall(), tf.keras.metrics.AUC(name='auc')]
    )
    return model

# Преобразуем все признаки с preprocessor
X_proc = preprocessor.fit_transform(X)
input_dim = X_proc.shape[1]

from sklearn.model_selection import ParameterGrid

# Гиперпараметры
param_grid = {
    'hidden_units': [[64,32], [128,64], [128,64,32]],
    'lr': [1e-3, 1e-4, 5e-4],
    'dropout': [0.2, 0.3, 0.4],
    'batch_size': [32],
    'epochs': [50]
}

cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=RANDOM_SEED)

results = []

for params in ParameterGrid(param_grid):
    fold_metrics = []
    print(f"Testing configuration: {params}")
    for fold, (train_idx, val_idx) in enumerate(cv.split(X_proc, y), 1):
        X_train_fold, X_val_fold = X_proc[train_idx], X_proc[val_idx]
        y_train_fold, y_val_fold = y.iloc[train_idx], y.iloc[val_idx]
        
        # TensorBoard log dir для fold
        log_dir = os.path.join(tempfile.gettempdir(), "tf_logs",
                               datetime.datetime.now().strftime("%Y%m%d-%H%M%S") + f"_fold{fold}")
        os.makedirs(log_dir, exist_ok=True)
        
        tb_cb = callbacks.TensorBoard(log_dir=log_dir, histogram_freq=0)
        es_cb = callbacks.EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)
        
        model = build_model(hidden_units=params['hidden_units'], lr=params['lr'], dropout=params['dropout'])
        history = model.fit(
            X_train_fold, y_train_fold,
            validation_data=(X_val_fold, y_val_fold),
            batch_size=params['batch_size'],
            epochs=params['epochs'],
            class_weight=class_weights,
            callbacks=[tb_cb, es_cb],
            verbose=0
        )
        
        metrics = model.evaluate(X_val_fold, y_val_fold, verbose=0)
        fold_metrics.append(metrics[1])  # accuracy
    mean_acc = np.mean(fold_metrics)
    results.append({'params': params, 'cv_accuracy': mean_acc})
    print(f"Mean CV accuracy: {mean_acc:.4f}\n")

results_df = pd.DataFrame(results).sort_values(by='cv_accuracy', ascending=False)
display(results_df.head(5))

best_params = results_df.iloc[0]['params']
print("Best configuration:", best_params)

final_model = build_model(hidden_units=best_params['hidden_units'],
                          lr=best_params['lr'],
                          dropout=best_params['dropout'])

final_model.fit(
    X_proc, y,
    batch_size=best_params['batch_size'],
    epochs=best_params['epochs'],
    class_weight=class_weights,
    verbose=1
)

from sklearn.metrics import accuracy_score

# Предсказание final NN
y_pred_nn = (final_model.predict(X_proc).ravel() > 0.5).astype(int)
nn_acc = accuracy_score(y, y_pred_nn)
print("Final NN accuracy:", nn_acc)
```
